```python
## Plot a Normal Distribution

import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

mean = 0
variance = 1
standard_deviation = math.sqrt(variance)

# Generate the PDF
x = np.linspace(mean - 3*standard_deviation, mean + 3*standard_deviation, 100)
y = stats.norm.pdf(x, mean, standard_deviation)

# Set up the plot
plt.plot(x, y)

# Set the plot title
plt.title("Normal Distribution")

# Save the plot as a PNG file
plt.savefig('normal_plot.png', dpi=100)

# Show the plot
plt.show()
```

```python
## Plot Standard Deviations from the Mean - Normal Distribution

import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

mean = 0
variance = 1
standard_deviation = math.sqrt(variance)

# Generate the PDF
x = np.linspace(mean - 3*standard_deviation, mean + 3*standard_deviation, 100)
y = stats.norm.pdf(x, mean, standard_deviation)

# Plot the PDF
plt.plot(x, y, label='Normal Distribution')

# Shade the area within one standard deviation of the mean
plt.fill_between(x, y, where=(x >= mean - standard_deviation) & (x <= mean + standard_deviation), color='midnightblue', alpha=0.5, label='68.2% observations')

# Add a label for the percentage within one standard deviation
plt.text(mean - 0.5*standard_deviation, 0.14, '34.1%', ha='center', va='center', fontsize=12)
plt.text(mean + 0.5*standard_deviation, 0.14, '34.1%', ha='center', va='center', fontsize=12)

# Shade the area within two standard deviations of the mean
plt.fill_between(x, y, where=(x >= mean - 2*standard_deviation) & (x <= mean + 2*standard_deviation), color='darkblue', alpha=0.5, label='95.4% observations')

# Add a label for the percentage within two standard deviations
plt.text(mean - 1.5*standard_deviation, 0.05, '13.6%', ha='center', va='center', fontsize=12)
plt.text(mean + 1.5*standard_deviation, 0.05, '13.6%', ha='center', va='center', fontsize=12)

# Shade the area within three standard deviations of the mean
plt.fill_between(x, y, where=(x >= mean - 3*standard_deviation) & (x <= mean + 3*standard_deviation), color='lightskyblue', alpha=0.5, label='99.7% observations')

# Add a label for the percentage within two standard deviations
plt.text(mean - 2.5*standard_deviation, 0.01,'2.1%', ha='center', va='center', fontsize=12)
plt.text(mean + 2.5*standard_deviation, 0.01, '2.1%', ha='center', va='center', fontsize=12)


# Add a legend
plt.legend(loc='upper right')
plt.xlabel('Standard Deviations from the Mean')
plt.ylabel('Density')

# Save the plot as a PNG file
plt.savefig('normal_stds.png', dpi=300)

# Show the plot
plt.show()

```
