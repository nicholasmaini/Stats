```python
## Plot a Normal Distribution

import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

mean = 0
variance = 1
standard_deviation = math.sqrt(variance)

# Generate the PDF
x = np.linspace(mean - 3*standard_deviation, mean + 3*standard_deviation, 100)
y = stats.norm.pdf(x, mean, standard_deviation)

# Set up the plot
plt.plot(x, y)

# Set the plot title
plt.title("Normal Distribution")

# Save the plot as a PNG file
plt.savefig('normal_plot.png', dpi=100)

# Show the plot
plt.show()
```

```python
## Plot Skewness - using a Dataframe - Multi-plot (Negative, Zero, Positive) - with Legend

import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
import pandas as pd
import seaborn as sns

# Set theme with Seaborn
sns.set(style='darkgrid', palette='deep', font='sans-serif')

# Set up the plot
fig, axs = plt.subplots(1, 3, figsize=(20, 3))

# Create a dataframe with the skew values
df = pd.DataFrame({'skew': [-5, 0, 5]})

# Loop through the skew values and generate the PDFs
for i, row in df.iterrows():
    skew = row['skew']
    x = np.linspace(-3, 3, 100)
    pdf = stats.skewnorm.pdf(x, skew)

    # Plot the PDFs
    axs[i].plot(x, pdf, label=f'Skew = {skew}')
    axs[i].legend()
    
    if skew < 0:
        axs[i].set_title('Negative Skew')
    elif skew == 0:
        axs[i].set_title('Normal (Zero Skew)')
    else:
        axs[i].set_title('Positive Skew')

    print((f'Skew = {skew}'))

# Save the plot as a PNG file
plt.savefig('skew_multiplot.png', dpi=300)

# Show the plot
plt.show()
```

```python
## Plot Kurtosis - using a Dataframe - Multi-plot (Platykurtic, Mesokurtic, Leptokurtic) - with Legend

from numpy import round

# Set up the plot
fig, ax = plt.subplots(1, 3, figsize=(20, 3))

# Generate the PDFs and plot them
xs = np.linspace(-5, 5, 100)
for i, dist in enumerate([stats.laplace, stats.norm, stats.cosine]):
    pdf = dist.pdf(xs)
    kurtosis = round(dist.stats(moments='k'), 1)
    ax[i].plot(xs, pdf, label=f'Kurtosis = {kurtosis}')
    ax[i].legend()
        
# Save the plot as a PNG file
plt.savefig('kurtosis_multiplot.png', dpi=300)

# Show the plot
plt.show()
```

```python
# Plot Skewness - using a Dataframe - Single Chart (Negative, Zero, Positive)

import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
import pandas as pd

# Set up the plot
fig, ax = plt.subplots(figsize=(10, 3))

# Create a dataframe with the skew values and labels
df = pd.DataFrame({'skew': [-5, 0, 5], 'label': ['Negative Skew', 'Zero Skew (Normal)', 'Positive Skew']})

# Generate the PDFs and plot them
x = np.linspace(-3, 3, 100)
for i, row in df.iterrows():
    skew = row['skew']
    label = row['label']
    pdf = stats.skewnorm.pdf(x, skew)
    ax.plot(x, pdf, label=label)

# Add a legend
ax.legend()

# Save the plot as a PNG file
plt.savefig('skew_plot.png', dpi=300)

# Show the plot
plt.show()
```

```python
# Plot Kurtosis - with a Dataframe - Single Chart (Platykurtic, Mesokurtic, Leptokurtic)

import matplotlib.pyplot as plt
import numpy as np
import scipy.stats
from numpy import round
import pandas as pd

# Set up the plot
fig, ax = plt.subplots(figsize=(10, 3))

# Create a dataframe with the kurtosis distribution functions and labels
df = pd.DataFrame({'dist': [stats.laplace, stats.norm, stats.cosine], 'label': ['Leptokurtic', 'Mesokurtic (Normal)', 'Platykurtic']})

# Generate the PDFs and plot them
xs = np.linspace(-5, 5, 100)
for i, row in df.iterrows():
    dist = row['dist']
    label = row['label']
    pdf = dist.pdf(xs)
    ax.plot(xs, pdf, label=label)
    print(f'{label}: {round(dist.stats(moments="k"), 1)}')

# Add a legend
ax.legend()

# Save the plot as a PNG file
plt.savefig('kurtosis_plot.png', dpi=300)

# Show the plot
plt.show()
```
